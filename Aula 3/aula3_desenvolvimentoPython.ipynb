```python
# ğŸ“Œ Conversor de Unidades
# Este programa converte diferentes unidades de medida:
# - Metros â†” QuilÃ´metros
# - QuilÃ´metros â†” Milhas
# - Celsius â†” Fahrenheit

def metros_para_quilometros(metros):
    return metros / 1000

def quilometros_para_metros(quilometros):
    return quilometros * 1000

def quilometros_para_milhas(quilometros):
    return quilometros * 0.621371

def milhas_para_quilometros(milhas):
    return milhas / 0.621371

def celsius_para_fahrenheit(celsius):
    return (celsius * 9/5) + 32

def fahrenheit_para_celsius(fahrenheit):
    return (fahrenheit - 32) * 5/9


# ğŸ“Œ Menu Interativo
def menu():
    while True:
        print("\n=== Conversor de Unidades ===")
        print("1. Metros â†’ QuilÃ´metros")
        print("2. QuilÃ´metros â†’ Metros")
        print("3. QuilÃ´metros â†’ Milhas")
        print("4. Milhas â†’ QuilÃ´metros")
        print("5. Celsius â†’ Fahrenheit")
        print("6. Fahrenheit â†’ Celsius")
        print("0. Sair")

        opcao = input("Escolha uma opÃ§Ã£o: ")

        if opcao == "0":
            print("Encerrando o programa. AtÃ© logo! ğŸ‘‹")
            break

        valor = float(input("Digite o valor a ser convertido: "))

        if opcao == "1":
            print(f"{valor} metros = {metros_para_quilometros(valor)} quilÃ´metros")
        elif opcao == "2":
            print(f"{valor} quilÃ´metros = {quilometros_para_metros(valor)} metros")
        elif opcao == "3":
            print(f"{valor} quilÃ´metros = {quilometros_para_milhas(valor)} milhas")
        elif opcao == "4":
            print(f"{valor} milhas = {milhas_para_quilometros(valor)} quilÃ´metros")
        elif opcao == "5":
            print(f"{valor} Â°C = {celsius_para_fahrenheit(valor)} Â°F")
        elif opcao == "6":
            print(f"{valor} Â°F = {fahrenheit_para_celsius(valor):.2f} Â°C")
        else:
            print("âš ï¸ OpÃ§Ã£o invÃ¡lida, tente novamente.")

# Executar o menu
menu()
```

```python
# Conversor de Unidades - VersÃ£o Completa e Refatorada

# ---------------- FunÃ§Ãµes de ConversÃ£o ----------------
def metros_para_quilometros(m): return m / 1000
def quilometros_para_metros(km): return km * 1000
def quilometros_para_milhas(km): return km * 0.621371
def milhas_para_quilometros(mi): return mi / 0.621371

def celsius_para_fahrenheit(c): return (c * 9/5) + 32
def fahrenheit_para_celsius(f): return (f - 32) * 5/9

def hectare_para_m2(h): return h * 10000
def m2_para_hectare(m2): return m2 / 10000
def acre_para_m2(a): return a * 4046.86
def m2_para_acre(m2): return m2 / 4046.86

def milha_nautica_para_km(mn): return mn * 1.852
def km_para_milha_nautica(km): return km / 1.852
def pes_para_km(pes): return pes * 0.0003048
def km_para_pes(km): return km / 0.0003048


# ---------------- DicionÃ¡rio de OpÃ§Ãµes ----------------
opcoes = {
    "1": ("Metros â†’ QuilÃ´metros", lambda v: f"{v} m = {metros_para_quilometros(v):.4f} km"),
    "2": ("QuilÃ´metros â†’ Metros", lambda v: f"{v} km = {quilometros_para_metros(v):.2f} m"),
    "3": ("QuilÃ´metros â†’ Milhas", lambda v: f"{v} km = {quilometros_para_milhas(v):.4f} milhas"),
    "4": ("Milhas â†’ QuilÃ´metros", lambda v: f"{v} mi = {milhas_para_quilometros(v):.4f} km"),
    "5": ("Celsius â†’ Fahrenheit", lambda v: f"{v} Â°C = {celsius_para_fahrenheit(v):.2f} Â°F"),
    "6": ("Fahrenheit â†’ Celsius", lambda v: f"{v} Â°F = {fahrenheit_para_celsius(v):.2f} Â°C"),
    "7": ("Hectare â†’ MetrosÂ²", lambda v: f"{v} ha = {hectare_para_m2(v):,.2f} mÂ²"),
    "8": ("MetrosÂ² â†’ Hectare", lambda v: f"{v} mÂ² = {m2_para_hectare(v):.4f} ha"),
    "9": ("Acre â†’ MetrosÂ²", lambda v: f"{v} acres = {acre_para_m2(v):,.2f} mÂ²"),
    "10": ("MetrosÂ² â†’ Acre", lambda v: f"{v} mÂ² = {m2_para_acre(v):.4f} acres"),
    "11": ("Milha NÃ¡utica â†’ QuilÃ´metros", lambda v: f"{v} MN = {milha_nautica_para_km(v):.4f} km"),
    "12": ("QuilÃ´metros â†’ Milha NÃ¡utica", lambda v: f"{v} km = {km_para_milha_nautica(v):.4f} MN"),
    "13": ("PÃ©s â†’ QuilÃ´metros", lambda v: f"{v} pÃ©s = {pes_para_km(v):.6f} km"),
    "14": ("QuilÃ´metros â†’ PÃ©s", lambda v: f"{v} km = {km_para_pes(v):.2f} pÃ©s"),
}


# ---------------- Menu Interativo ----------------
def menu():
    while True:
        print("\n=== Conversor de Unidades ===")
        for chave, (descricao, _) in opcoes.items():
            print(f"{chave}. {descricao}")
        print("0. Sair")

        escolha = input("\nEscolha uma opÃ§Ã£o: ")

        if escolha == "0":
            print("Saindo do programa... ğŸ‘‹")
            break

        if escolha not in opcoes:
            print("âŒ OpÃ§Ã£o invÃ¡lida, tente novamente.")
            continue

        try:
            valor = float(input("Digite o valor para conversÃ£o: "))
            resultado = opcoes[escolha][1](valor)
            print("âœ… Resultado:", resultado)
        except ValueError:
            print("âš ï¸ Digite um nÃºmero vÃ¡lido!")

# ---------------- Executar ----------------
menu()
```

```python
# gerador_senhas.py
import secrets
import string
import os

ARQUIVO_SENHAS = "senhas_geradas.txt"

def carregar_senhas_existentes():
    """Carrega um conjunto com as senhas jÃ¡ geradas (persistÃªncia simples)."""
    if not os.path.exists(ARQUIVO_SENHAS):
        return set()
    with open(ARQUIVO_SENHAS, "r", encoding="utf-8") as f:
        return set(l.strip() for l in f if l.strip())

def salvar_senha(senha: str):
    with open(ARQUIVO_SENHAS, "a", encoding="utf-8") as f:
        f.write(senha + "\n")

def montar_charset(usamin, usamaj, usanum, usasym):
    charset = ""
    if usamin: charset += string.ascii_lowercase
    if usamaj: charset += string.ascii_uppercase
    if usanum: charset += string.digits
    if usasym:
        # Conjunto enxuto de sÃ­mbolos comuns (pode ampliar se quiser)
        charset += "!@#$%^&*()-_=+[]{};:,.?/\\|~"
    return charset

def gerar_unica(tam: int, charset: str, ja_usadas: set, max_tentativas: int = 10000) -> str:
    """Gera uma senha Ãºnica (nÃ£o repetida). Tenta atÃ© max_tentativas vezes."""
    if not charset:
        raise ValueError("Nenhum conjunto de caracteres selecionado.")
    for _ in range(max_tentativas):
        senha = "".join(secrets.choice(charset) for _ in range(tam))
        if senha not in ja_usadas:
            return senha
    raise RuntimeError("NÃ£o foi possÃ­vel gerar uma senha Ãºnica com os parÃ¢metros atuais.")

def garantir_requisitos_basicos(senha: str, usamin, usamaj, usanum, usasym) -> bool:
    """Opcional: assegura que a senha contenha pelo menos um de cada tipo selecionado."""
    if usamin and not any(c.islower() for c in senha): return False
    if usamaj and not any(c.isupper() for c in senha): return False
    if usanum and not any(c.isdigit() for c in senha): return False
    if usasym and not any(c in "!@#$%^&*()-_=+[]{};:,.?/\\|~" for c in senha): return False
    return True

def gerar_senha_com_requisitos(tam, usamin, usamaj, usanum, usasym, ja_usadas):
    charset = montar_charset(usamin, usamaj, usanum, usasym)
    # Garantir que o tamanho comporte os requisitos mÃ­nimos
    requisitos = sum([usamin, usamaj, usanum, usasym])
    if requisitos > tam:
        raise ValueError(f"O tamanho deve ser â‰¥ ao nÃºmero de tipos selecionados ({requisitos}).")

    # Gera atÃ© atender requisitos e unicidade
    for _ in range(20000):
        senha = gerar_unica(tam, charset, ja_usadas, max_tentativas=1000)
        if garantir_requisitos_basicos(senha, usamin, usamaj, usanum, usasym):
            return senha
    raise RuntimeError("NÃ£o foi possÃ­vel gerar uma senha Ãºnica que atenda aos requisitos.")

def menu():
    print("=== Gerador de Senhas (sem repetiÃ§Ã£o) ===")
    print("As senhas geradas sÃ£o salvas em 'senhas_geradas.txt' para evitar duplicatas.")
    ja_usadas = carregar_senhas_existentes()

    try:
        tam = int(input("Comprimento da senha (ex: 12): ").strip())
        if tam <= 0:
            raise ValueError

        usamin = input("Incluir letras minÃºsculas? (s/n): ").strip().lower() == "s"
        usamaj = input("Incluir letras MAIÃšSCULAS? (s/n): ").strip().lower() == "s"
        usanum = input("Incluir nÃºmeros? (s/n): ").strip().lower() == "s"
        usasym = input("Incluir sÃ­mbolos? (s/n): ").strip().lower() == "s"

        if not any([usamin, usamaj, usanum, usasym]):
            print("VocÃª precisa selecionar pelo menos um tipo de caractere.")
            return

        senha = gerar_senha_com_requisitos(tam, usamin, usamaj, usanum, usasym, ja_usadas)
        salvar_senha(senha)
        print("\nâœ… Senha gerada (Ãºnica):", senha)

    except ValueError:
        print("Entrada invÃ¡lida. Tente novamente com nÃºmeros vÃ¡lidos.")
    except RuntimeError as e:
        print("Erro:", e)

if __name__ == "__main__":
    menu()
```

```python
import random
import string
from colorama import Fore, Style, init

# Inicializar o colorama
init(autoreset=True)

# Conjunto global para armazenar senhas Ãºnicas
senhas_geradas = set()

def gerar_senha(tamanho=12, usar_maiusculas=True, usar_minusculas=True, usar_numeros=True, usar_simbolos=True):
    """Gera uma senha Ãºnica com base nas opÃ§Ãµes fornecidas."""
    caracteres = ""
    if usar_maiusculas:
        caracteres += string.ascii_uppercase
    if usar_minusculas:
        caracteres += string.ascii_lowercase
    if usar_numeros:
        caracteres += string.digits
    if usar_simbolos:
        caracteres += string.punctuation

    if not caracteres:
        raise ValueError("âš ï¸ Ã‰ necessÃ¡rio escolher pelo menos um tipo de caractere!")

    # Gera senhas atÃ© achar uma nova
    while True:
        senha = "".join(random.sample(caracteres, tamanho)) if tamanho <= len(caracteres) else "".join(random.choices(caracteres, k=tamanho))
        if senha not in senhas_geradas:
            senhas_geradas.add(senha)
            return senha

def menu():
    print(Fore.CYAN + Style.BRIGHT + "\nğŸ” GERADOR DE SENHAS ALEATÃ“RIAS ğŸ”\n")

    try:
        tamanho = int(input(Fore.YELLOW + "Digite o comprimento da senha (ex: 12): "))
    except ValueError:
        print(Fore.RED + "âš ï¸ Entrada invÃ¡lida! Usando valor padrÃ£o (12).")
        tamanho = 12

    usar_maiusculas = input("Incluir letras maiÃºsculas? (s/n): ").lower() == "s"
    usar_minusculas = input("Incluir letras minÃºsculas? (s/n): ").lower() == "s"
    usar_numeros = input("Incluir nÃºmeros? (s/n): ").lower() == "s"
    usar_simbolos = input("Incluir sÃ­mbolos? (s/n): ").lower() == "s"

    senha = gerar_senha(
        tamanho,
        usar_maiusculas,
        usar_minusculas,
        usar_numeros,
        usar_simbolos
    )

    print(Fore.GREEN + Style.BRIGHT + f"\nâœ… Sua senha gerada Ã©: {senha}\n")

if __name__ == "__main__":
    while True:
        menu()
        continuar = input(Fore.MAGENTA + "Deseja gerar outra senha? (s/n): ").lower()
        if continuar != "s":
            print(Fore.CYAN + "ğŸ‘‹ Encerrando o gerador de senhas. AtÃ© mais!")
            break
```
